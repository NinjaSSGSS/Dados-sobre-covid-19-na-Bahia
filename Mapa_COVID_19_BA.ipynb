```
import pandas as pd
import folium
```

```
arquivo = pd.read_csv('/content/drive/My Drive/Git Hub/covid19br/cases-brazil-cities.csv')
arquivo = arquivo.query('state == "BA"')
arquivo.reset_index(inplace=True)
arquivo.drop(columns=['country', 'ibgeID', 'state', 'deaths_per_100k_inhabitants','totalCases_per_100k_inhabitants', 'deaths_by_totalCases', 'index'], inplace=True)
arquivo.insert(loc=3, column='Lat', value=0)
arquivo.insert(loc=4, column='Long', value=0)
coord = pd.read_excel('/content/drive/My Drive/Mapas/Lat&Lon_Mun_da_Ba.xlsx')
coord.drop(columns=['Unnamed: 0', 'NM_UF'], inplace=True)
coord = coord[['NM_MUNICIPIO', 'LAT', 'LONG']]
```

```
arquivo_len = arquivo.shape[0]
coord_len = coord.shape[0]
for x in range(arquivo_len):
  for y in range(coord_len):
    if arquivo['city'][x][0:len(arquivo['city'][x])-3].lower() == coord['NM_MUNICIPIO'][y].lower():
      arquivo['Lat'][x] = str(coord['LAT'][y])
      arquivo['Long'][x] = str(coord['LONG'][y])
arquivo['Lat'] = pd.to_numeric(arquivo['Lat'])
arquivo['Long'] = pd.to_numeric(arquivo['Long'])
```

```
mapa = folium.Map(location=[-12.216048, -41.142351],zoom_start=5.5, min_zoom=5.5, max_zoom = 8, prefer_canvas=True)
```

```
sum_casos = arquivo['totalCases'].sum()
total_radius = 100
```

```
for x in range(arquivo_len):
  lat = arquivo['Lat'][x]
  longt = arquivo['Long'][x]
  locat = [lat, longt]
  nome = arquivo['city'][x]
  totalCases = arquivo['totalCases'][x]
  radius = (totalCases*total_radius)/sum_casos
  if totalCases != 0:
    if nome != 'Salvador/BA' and totalCases < 16:
      folium.CircleMarker(location=locat, color = '#ff0000', radius = radius*25, fill=True, fill_color = '#ff0000', tooltip='{}, {} casos'.format(nome, totalCases)).add_to(mapa)
    elif nome != 'Salvador/BA' and totalCases > 16:
      folium.CircleMarker(location=locat, color = '#ff0000', radius = radius*5, fill=True, fill_color = '#ff0000', tooltip='{}, {} casos'.format(nome, totalCases)).add_to(mapa)
    else:
      folium.CircleMarker(location=locat, color = '#ff0000', radius = (radius), fill=True, fill_color = '#ff0000', tooltip='{}, {} casos'.format(nome, totalCases)).add_to(mapa)
```
